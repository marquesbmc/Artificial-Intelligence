```python
import pandas as pd
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Etapa 3: Ler o arquivo CSV
df = pd.read_csv(io.StringIO(uploaded['dados_treinamento-2k-32.csv'].decode('utf-8')))
```

```python
df.head()
```

```python
df.drop(columns=['Episodio'], inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

# Suponha que você tenha um DataFrame chamado df com os dados que deseja plotar.

# Gráfico 1: Recompensa
plt.figure(figsize=(8, 4))
plt.plot(df['Recompensa'], marker='o')
plt.title('Gráfico de Recompensa')
plt.xlabel('Índice')
plt.ylabel('Recompensa')
plt.show()

# Gráfico 2: Comprimento do Episódio
plt.figure(figsize=(8, 4))
plt.plot(df['Comprimento do Episodio'], marker='o')
plt.title('Gráfico de Comprimento do Episódio')
plt.xlabel('Índice')
plt.ylabel('Comprimento do Episódio')
plt.show()

# Gráfico 3: Loss
plt.figure(figsize=(8, 4))
plt.plot(df['Loss'], color='red', marker='o')
plt.title('Gráfico de Loss')
plt.xlabel('Índice')
plt.ylabel('Loss')
plt.show()

# Gráfico 4: Epsilon
plt.figure(figsize=(8, 4))
plt.plot(df['epsilon'], color='green', marker='o')
plt.title('Gráfico de Epsilon')
plt.xlabel('Índice')
plt.ylabel('Epsilon')
plt.show()
```

```python
import numpy as np

# Calculando a média de cada 50 linhas
novo_df = df.groupby(np.arange(len(df)) // 50).mean().reset_index(drop=True)
```

```python
novo_df.head()
```

```python
import matplotlib.pyplot as plt

# Suponha que você tenha um DataFrame chamado df com os dados que deseja plotar.

# Gráfico 1: Recompensa
plt.figure(figsize=(8, 4))
plt.plot(novo_df['Recompensa'], marker='o')
plt.title('Gráfico de Recompensa')
plt.xlabel('Índice')
plt.ylabel('Recompensa')
plt.show()

# Gráfico 2: Comprimento do Episódio
plt.figure(figsize=(8, 4))
plt.plot(novo_df['Comprimento do Episodio'], marker='o')
plt.title('Gráfico de Comprimento do Episódio')
plt.xlabel('Índice')
plt.ylabel('Comprimento do Episódio')
plt.show()

# Gráfico 3: Loss
plt.figure(figsize=(8, 4))
plt.plot(novo_df['Loss'], color='red', marker='o')
plt.title('Gráfico de Loss')
plt.xlabel('Índice')
plt.ylabel('Loss')
plt.show()

# Gráfico 4: Epsilon
plt.figure(figsize=(8, 4))
plt.plot(novo_df['epsilon'], color='green', marker='o')
plt.title('Gráfico de Epsilon')
plt.xlabel('Índice')
plt.ylabel('Epsilon')
plt.show()
```
